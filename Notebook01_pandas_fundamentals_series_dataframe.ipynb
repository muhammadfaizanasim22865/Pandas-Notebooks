# Intro to Pandas

Pandas is a powerful Python library used for handling and analyzing structured data. It helps us work with tables, rows, and columns
 
 similar to Excel sheets inside Python. In Machine Learning and Data Science, Pandas is mainly used to 
 
 load, clean, organize, filter, and analyze datasets before giving them to ML models.

### importing pandas

```python
import pandas as pd
```

### What is a Series?

A Series is a one-dimensional labeled array. Think of it like a single column in a spreadsheet or a single hallway 

of apartments where each door has a label and holds some data

### Creating a Basic Series

To create a Series, you first need to import Pandas and then use the Series constructor (note the capital 'S').

***From a List  :***

```python
data = [100, 102, 104]
seriesA = pd.Series(data)
print(seriesA)
```

***Using Dictionaries***

When you pass a Python Dictionary into the Series constructor, Pandas automatically uses

 the dictionary keys as labels and the values as the data

```python
calories = {"day1": 1750, "day2": 2100, "day3": 1700}
seriesB = pd.Series(calories)
print(seriesB)
```

### Customizing the Index

You can replace the default numeric index (0, 1, 2...) with custom labels using the index argument

```python
seriesA = pd.Series(data, index=["A", "B", "C"])
print(seriesA)
```

### Accessing and Updating Values

***.loc[ ] (Label-based)  :*** 

Access a value by its custom label.

```python
print(seriesA.loc["A"])
```

***.iloc[ ] (Integer-based)  :***

 Access a value by its position (0-indexed)

```python
print(seriesB.iloc[0])
```

***Updating  :*** 

To change a value, simply assign a new one to a location

```python
seriesA.loc["C"] = 200
print(seriesA)
```

### Filtering Data

You can filter a Series by applying a condition within the subscript operator []

```python
print(seriesA[seriesA > 100])
```

###  What is a DataFrame?

A DataFrame is a two-dimensional, tabular data structure with rows and columns, similar to an Excel spreadsheet

### Creating a DataFrame

***From a Dictionary  :***

```python
data = {
    "name": ["Spongebob", "Patrick", "Squidward"],
    "age": [30, 35, 50]
}
df = pd.DataFrame(data)
print(df)
```

***Adding a Custom Index  :***
 
 By default, rows are labeled 0, 1, 2.... You can specify your own labels 




```python
df = pd.DataFrame(data, index=["Employee A", "Employee B", "Employee C"])
print(df)
```

### Adding a New Column

You can add a new column by assigning a list of values to a new column name 

The length of the list must match the number of rows

```python
df["job"] = ["cook","musician","doctor"]
print(df)
```

### Adding New Rows

The recommended way to add rows is by creating a new temporary DataFrame and using pd.concat( ) (as the old .append() method is deprecated) 

Create a new data structure for the row(s).

Convert it to a DataFrame.

Concatenate it with the original.

```python
new_row_data =[{"name":"sandy", "age":31, "job":None}]
new_row_df = pd.DataFrame(new_row_data,index=["Employee D"]) 
df = pd.concat([df,new_row_df])

print(df)
```

Core EDA function

#### Why?

Provides statistical summary of numerical columns.

Shows  :

- mean

- std

- min

- max

- quartiles


Used for  :

- understanding distributions

- detecting unusual values

- summarizing numerical data

#### `describe(include="all")`

Generates a statistical summary for both numerical and categorical columns.

NaN appears where a statistic does not apply to that data type (e.g., mean for strings).

#### Examples




```python
print(df.describe())
```

```python
print(df.describe(include="all"))
```

#### 1️⃣ The Setup

Create a DataFrame named `payroll_df` representing the employees of the Krusty Krab.

Use the following data:

| Name | Base Salary | Department |
|---|---|---|
| Spongebob | 500 | Kitchen |
| Patrick | 400 | Kitchen |
| Squidward | 600 | Counter |

Use custom index labels:

ID-01
ID-02
ID-03

#### Solution

```python
payroll_df = pd.DataFrame({"name": ["Spongebob", "Patrick", "Squidward"],
                            "Base salary": [500, 400, 600],
                              "Department": ["Kitchen", "Kitchen", "Counter"]
                              }
                              , index=["ID-01", "ID-02", "ID-03"])
print(payroll_df)
```

#### Add Bonus Column

```python
payroll_df["bonus"] = [50,50,50]
print(payroll_df)
```

#### Add New Hire

```python
new_hire = pd.DataFrame({"name":"Sandy", "Base salary":[700], "Department":["Science"], "bonus":[100]}, index=["ID-04"])
payroll_df = pd.concat([payroll_df, new_hire])
print(payroll_df)
```

#### Update Value

```python
payroll_df.loc["ID-03", "Base salary"] = 650
print(payroll_df)
```

#### Filter High Earners

```python
high_earners = payroll_df[payroll_df["Base salary"]>550]
print(high_earners)
```

#### Average Salary

```python
average_Salary = payroll_df["Base salary"].mean()
print(average_Salary)
```
