# 📂 Loading Datasets

Real-world datasets are usually stored in files such as:

- CSV

- JSON

- Excel

Pandas can load these files into DataFrames for analysis.

## 📄 Loading CSV Files

### `read_csv()`

Used to load CSV (Comma Separated Values) files.

### Example


```python
import pandas as pd
df = pd.read_csv("student_data.csv")
print(df)
```

## `to_string( )`

`to_string( )` converts a pandas DataFrame or Series into formatted text.

```python
print(df.to_string())
```

## 🌐 Loading JSON Files

### `read_json()`

Used to load JSON datasets.

### Example

```python
print(pd.read_json("student.json"))
```

## column Selection & Filtering in Pandas

 column selection is one of the most important parts of Pandas.

It allows us to:

- access specific data

- filter useful information

- analyze datasets

- prepare data for Machine Learning

This is heavily used in:

- Data Analysis

- Exploratory Data Analysis (EDA)

- Data Cleaning

- Machine Learning preprocessing


### Selecting a Single Column

Selecting one column returns a **Series**.

### Example

```python
print(df["school"])
```

### Selecting Multiple Columns

Selecting multiple columns returns a **DataFrame**.

### Example


```python
print(df[["school", "age"]])
```

## Filtering Data

Filtering is used to display only the rows that match a condition.

This is one of the MOST important Pandas skills.


### Single Condition Filtering


```python
print(df[df["age"] > 15])
```

### Multiple Conditions using `&` (AND)

`&` means BOTH conditions must be true.

### Example


```python
guy_students = df[
    (df["sex"] == "M") & (df["age"] > 20)
]

print(guy_students)
```


### Important

Each condition must be inside parentheses `( )`.


### Multiple Conditions using `|` (OR)

`|` means ANY ONE condition can be true.

### Example

```python
mature = df[(df["sex"]=="M") | (df["age"] > 20)]
print(mature)
```

## Sorting Data

### sort_values( )

Used to sort data based on a column.

### Example

```python
print(df.sort_values("age", ascending=False))
```

## Value Analysis

### value_counts( )

Counts frequency of each unique value in a column.

### Example

```python
print(df["school"].value_counts())
```

### unique( )

Returns all unique values in a column.

### Example

```python
print(df["Mjob"].unique())
```

### nunique( )

Returns number of unique values in a column.

### Example

```python
print(df["Fjob"].nunique())
```

## Grouping

### groupby( )

Groups data based on a column.

### Example

```python
for name in df.groupby("school"):
    print(name)
```

## Aggregation with groupby( )

- Splits data into groups

- Applies calculations per group

- Very important for real-world analysis



```python
print(df.groupby("school").agg({"age":["mean","sum","max","min"]}))
```

```python
selected_columns = df[["school", "age", "sex"]]
print(selected_columns)
```

## 6️⃣ Grouping Practice

```python
grouping = df.groupby("school").agg({"absences" : ["mean","sum","count"]})
print(grouping)
```
